## Matplotlib
Matplotlib is a library for 2-dimensional plotting with Python.
Some Plots:
Plots
Histograms
Error charts
Power spectra
Bar charts
Scatter Plots
    


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4,4,5,])
plt.xlabel("Actual birth weight") # Set XLabel
plt.ylabel("Estimated birth weight") # Set Ylabel
plt.show()
```

## Formatting your Python Plot
A third argument will let you choose the color and the line type of the plot

```python
plt.plot([1,2,3,4,4,5],[5,6,2,10,15,12],"gs")
plt.xlabel("Actual birth weight") # Set XLabel
plt.ylabel("Estimated birth weight") # Set Ylabel
plt.show()
```

```python
t=np.arange(0,5,0.2)
plt.plot(t,t,'r--',t,t**3,'b^',t,t**2,'gs')
```

## Matplotlib Keyword Strings
Using a data keyword argument, we can produce Python plots with strings that relate to some variables. 

```python
data = {'a': np.arange(50),'c' : np.random.randint(0,50,50),'d':np.random.randn(50)}
data['b']=data['a']+10*np.random.randn(50)
data['d']=np.abs(data['d'])*100
data['e'] = data['a'] + data['b']
```

```python
plt.scatter('a','b',c = 'c',s='d',data = data)
```

## Categorical Variable

```python
names=["Dingos","Wild Cats","Tigers"]
values=[1,11,111]
plt.figure(1,figsize=(9,5)) # Size of plotting

plt.subplot(131) # First graph position
plt.bar(names,values)

plt.subplot(132) # Second graph position
plt.scatter(names,values)

plt.subplot(133) # Third graph position
plt.plot(names,values)

plt.suptitle('Varsity')
plt.show()
```

## Some Line Properties
1. Linewidth
2. Alpha : softening colors
3. Antialiased
4. Color or c
5 . Dashes
6. LineStyle or ls
7. Marker
8. MarkerEdgeColor
9. MarkerEdgeWidth
10. MarkerFacecolor and MarkerSize
11.  MarkEvery
12. Zorder

```python
plt.plot([1,2,3],[2,4,9],linewidth=4.0,alpha=5.5,antialiased=True,color='Chartreuse',dashes=[1,2,4,4])
```

```python
plt.plot([1,2,3],[2,4,9],linewidth=4.0,alpha=5.5,antialiased=True,color='Chartreuse',linestyle=':')
```

```python
plt.plot([1,2,3],[2,4,9],marker='+',markeredgecolor='brown',markeredgewidth=0.8,markerfacecolor='orange',markersize=13.0)
```
