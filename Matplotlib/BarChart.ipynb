## Histogram
A histogram is a graph that represents the way numerical data is represented. The input to it is a numerical variable, which it separates into bins on the x-axis.
This is a vector of numbers and can be a list or a DataFrame column

```python
import seaborn as sn
import pandas as pd
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
data = iris.data
features = iris.target_names
```

```python
dataset = pd.DataFrame(data[:,0:3],columns = features)
dataset.head(3)
```

```python
sn.distplot(dataset['setosa'])
```

```python
sn.distplot(dataset['setosa'],bins=25)
```

## Displaying Only The Histogram

```python
sn.distplot(a=dataset['setosa'],hist=True,kde=False,rug=False)
```

## Displaying Histogram, Rug, and Kernel Density

```python
sn.distplot(a=dataset['setosa'],hist=True,kde=True,rug=True)
```
